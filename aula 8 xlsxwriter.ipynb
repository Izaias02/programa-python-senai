exemplo 2: aplicando formatação 

```python
import xlsxwriter as xlsx
import os
caminho = 'C:\\Users\\FIC\\Documents\\izaias python\\'
arquivo = 'exemplo-2.xlsx'

# --------- criar um arquivo e uma aba

workbook = xlsx.Workbook(caminho + arquivo)
sheetDados = workbook.add_worksheet('dados') 

#------ modificando a cor de fundo

corFonte = workbook.add_format()
corFonte.set_font_color('blue')

# ------mudar cor de fundo

corfundo = workbook.add_format({'fg_color':'yellow',# cor preenchimento
                               'font_color':'gray',# cor da fonte
                                'bold':True, # negrito
                                'italic': True}) # aplicar bordas

#-------------- escrever nas celulas

sheetDados.write('A1', 'nome',corfundo)
sheetDados.write('A2', 'eduardo',corFonte)
sheetDados.write('A3', 'julia',corfundo)
sheetDados.write('A4', 'josé',corFonte)
sheetDados.write('A5', 'jacqueline',corfundo)

sheetDados.write('B1', 'idade',corfundo)
sheetDados.write('B2', '18',corFonte)
sheetDados.write('B3', '21',corfundo)
sheetDados.write('B4', '19',corFonte)
sheetDados.write('B5', '26',corfundo)

# ------fechar o documento e salvar
workbook.close()

os.startfile(caminho + arquivo)

```

```python
import xlsxwriter as xlsx
import os

caminho = 'C:\\Users\\FIC\\Documents\\izaias python\\'
arquivo = 'desafio-1.xlsx'

workbook = xlsx.Workbook(caminho + arquivo)
sheetformataçao = workbook.add_worksheet('formatação')
# formatação celulas coluna A

corcabeçalholinhas = workbook.add_format({
                                        'fg_color':'#F4D0C4',
                                        'font_color':'#1D3857',
                                        'bold': True,
                                        'border': True,
                                        })

# formatação valores

corvalores = workbook.add_format({
                                'num_format':'R$ #,##0'
                                })

# formatação dos totais 

cortotais = workbook.add_format({
                                'num_format':'R$ #,##0',
                                'fg_color':'#F4D0C4',
                                'font_color':'#1D3857',
                                'bold': True,
                                })
# formatação cabeçalho mesclados

corcabecalhosmesclados = workbook.add_format({
                                            'bold':True,
                                            'font_size': 18,
                                            'align':'center'
                                            })

corfonte = workbook.add_format()
corfonte.set_font_color('black')

corfundo1 = workbook.add_format({'fg_color':'green',})
corfundo2 = workbook.add_format({'fg_color':'yellow'})

sheetformataçao.merge_range('A1:C1', 'CONTROLE DE CUSTOS',corcabecalhosmesclados)

sheetformataçao.merge_range('A2:A3', 'itens',corcabecalhosmesclados)

sheetformataçao.write('A4', 'salario',corcabeçalholinhas)
sheetformataçao.write('A5', 'horas extras ',corcabeçalholinhas)
sheetformataçao.write('A6', 'previdencia social',corcabeçalholinhas)
sheetformataçao.write('A7', 'FGTS',corcabeçalholinhas)
sheetformataçao.write('A8', 'aluguel',corcabeçalholinhas)
sheetformataçao.write('A9', 'luz/agua/telefone',corcabeçalholinhas)
sheetformataçao.write('A10', 'mat. expediente',corcabeçalholinhas)
sheetformataçao.write('A11', 'outras despesas',corcabeçalholinhas)
sheetformataçao.write('A12', 'total',corcabecalhosmesclados)



sheetformataçao.merge_range('B2:C2', 'TOTAL 1° TRIMESTRE',corcabecalhosmesclados)
sheetformataçao.write('B3', 'previsao',corcabecalhosmesclados)
sheetformataçao.write('B4', 2000,corvalores)
sheetformataçao.write('B5', 3150,corvalores)
sheetformataçao.write('B6', 4300,corvalores)
sheetformataçao.write('B7', 5450,corvalores)
sheetformataçao.write('B8', 6600,corvalores)
sheetformataçao.write('B9', 7750,corvalores)
sheetformataçao.write('B10', 8900,corvalores)
sheetformataçao.write('B11', 10050,corvalores)




sheetformataçao.write('C3', 'gastos',corcabecalhosmesclados)
sheetformataçao.write('C4', 1550,corvalores)
sheetformataçao.write('C5', 2550,corvalores)
sheetformataçao.write('C6', 3550,corvalores)
sheetformataçao.write('C7', 4550,corvalores)
sheetformataçao.write('C8', 5550,corvalores)
sheetformataçao.write('C9', 6550,corvalores)
sheetformataçao.write('C10', 7550,corvalores)
sheetformataçao.write('C11', 8550,corvalores)

sheetformataçao.write_formula('B12', '=sum(B4:B11)',cortotais)
sheetformataçao.write_formula('C12', '=sum(C4:C11)',cortotais)

workbook.close()

os.startfile(caminho + arquivo)



# (adobe color) para buscar as cores em codigo para aplicar na formatação 
# 'num_format':'R$ #,##0'  para transformat um numero em moeda
```
